# Equations

## General Equations

$$
\tag{1}
Y = C + I + \Delta IN
$$

$$
\tag{2}
I = I_f + p_h\cdot I_h
$$

$$
\tag{3}
\omega = \overline\omega
$$

$$
W = \omega\cdot Y
\tag{4}
$$

$$
Y_K = {\frac{K_{t-1}}{\overline v}}
\tag{5}
$$

$$
u = \frac{Y}{Y_K}
\tag{6}
$$

$$
g_k = \frac{h\cdot u}{v}
\tag{7}
$$

$$
Z = I_h
\tag{8}
$$

$$
V = V_h + V_f + V_b = K
\tag{9}
$$

## Households

$$
YD = W + FD + \overline r_{m-1}\cdot M_{-1} - \overline r_{mo-1}\cdot MO
\tag{10}
$$

$$
S_h = YD - C
\tag{11}
$$

$$
\Delta MO = I_h\cdot p_h
\tag{12}
$$

$$
C = \alpha_1\cdot W + \alpha_2\cdot V_{h_{-1}}
\tag{13}
$$

$$
V_h = M + p_h\cdot H^D - MO
\tag{14}
$$
  

$$
CG = \Delta p_h\cdot H^D_{t-1} + p_{h_{t-1}}\cdot\Delta H^D
\tag{15}
$$

## Firms

$$
\Delta L = I_f + \Delta IN - FU
\tag{16}
$$

$$
FT = Y - W = FU+ FD
\tag{17}
$$

$$
FU = \gamma_F\cdot (FT - r_{L_{-1}}\cdot L_{-1})
\tag{18}
$$


$$
FD = (1 - \gamma_F)\cdot (FT - r_{L_{-1}}\cdot L_{-1})
\tag{19}
$$

$$
I_f = h\cdot Y
\tag{20}
$$

$$
\Delta h = 
h_{-1}\cdot \gamma_u\cdot (u - \overline u_n)
\tag{21}
$$

$$
V_f = K_f + p_h\cdot IN - L
\tag{22}
$$

$$
S_f = FU - I_f - \Delta IN
\tag{23}
$$

## Banks

$$
\Delta M = \Delta L + \Delta MO
\tag{24}
$$

$$
S_b = rl_{-1}\cdot L_{-1} + rmo_{-1}\cdot MO_{-1} - rm_{-1}\cdot M_{-1}
\tag{25}
$$

$$
r_l = r_m + spread_l
\tag{26}
$$

$$
r_{mo} = r_m + spread_{mo}
\tag{27}
$$

$$
V_b = L + MO - M
\tag{28}
$$

## Residential Investment


$$
\Delta IN = H^S - H^D
\tag{29}
$$

$$
I_h = \gamma_{0} - \gamma_{1}\cdot r_H
\tag{30}
$$

$$
\Delta H^{D} = I_h
\tag{31}
$$

$$
r_H = \left(\frac{1 + r_{MO}}{1 + \dot p_h}\right) -1
\tag{32}
$$

$$
\Delta p_h = -\gamma_{ph}\cdot\Delta IN
\tag{33}
$$

$$
\dot p_h = \frac{\Delta p_h}{p_{h_{-1}}} \hspace{3cm} p_h > 0
\tag{34}
$$


$$
H^S = H^D_e + \Delta IN^e
\tag{35}
$$

$$
H^D_e = H^D_{e_{t-1}} + \beta_H\cdot (H^D_{{t-1}} - H^D_{e_{t-1}})
\tag{36}
$$

$$
\Delta IN^e = IN_{t-1} + \beta_{IN}\cdot (IN^T - IN_{t-1})
\tag{37}
$$

$$
IN^T = \sigma^T\cdot (H^D_e)
\tag{38}
$$

$$
g_{I_h} = \frac{I_f}{H^D_{t-1}}
\tag{39}
$$

$$
g_{I_h}^e = g_{I_{h-1}} + \beta\cdot (g_{I_{h-1}}^e - g_{I_{h-1}})
\tag{40}
$$

## Loading libraries

```
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import seaborn as sns
from pysolve3.utils import ShockModel, SFCTable, SolveSFC
from pysolve3.model import Model
import sympy as sp
```

# Creating model function

```
def model():
  model = Model()
  model.set_var_default(0)
  
  model.var('C', desc='Consumption')
  model.var('CG', desc='Capital Gains')
  model.var('FD', desc='Distributed profits')
  model.var('FT', desc='Total Profits')
  model.var('FU', desc='Retained profits')
  model.var('gk', desc='Capital growth rate', default=0.01)
  model.var('g_Ih', desc='Residential investment growth rate (Demand)', default = 0.02)
  model.var('g_Ihe', desc='Expected residential investment growth rate')
  model.var('h', desc='Marginal propensity to invest (non-residential)', default=0.3) # Certo: 0.06
  model.var('H_S', desc='Houses supply', default=600)
  model.var('H_D', desc='Houses demand', default=250) # 150
  model.var('H_De', desc='Expected houses demand', default=350)
  model.var('IN', desc='Unsold houses (Inventories)', default=350) # 50
  model.var('INe', desc='Expected target Unsold houses (Inventories)')
  model.var('INT', desc='Target Unsold houses (Inventories)', default=70)
  model.var('I_t', desc='Investment', default = 120)
  model.var('I_f', desc='Non-residential investment', default = 100)
  model.var('I_h', desc='Residential investment', default=20)
  model.var('inf_ph', desc='Housing inflation', default=0.01)
  model.var('K_f', desc='Non-residential capital', default = 600)
  model.var('K', desc='Capital', default=900)
  model.var('L', desc='Loans')
  model.var('M', desc='Money deposits')
  model.var('MO', desc='Mortgages', default = 200)
  model.var('omega', desc='Wage-share', default=0.5)
  model.var('ph', desc='Housing prices', default=100)
  model.var('rl', desc='Interests rates on loans')
  model.var('rmo', desc='Interests rates on mortgages')
  model.var('rH', desc='Houses own interest rate')
  model.var('S_h', desc='Households savings')
  model.var('S_f', desc='Firms savings')
  model.var('S_b', desc='Banks savings')
  model.var('u', desc='Capacity utilization ratio', default=0.7)
  model.var('V_h', desc='Household net financial wealth')
  model.var('V_f', desc='Firms net financial wealth')
  model.var('V_b', desc='Banks net financial wealth')
  model.var('W', desc='Wages')
  model.var('Y', desc='GDP', default = 400)
  model.var('Yk', desc='Capacity', default = 500)
  model.var('YD', desc='Household disposable income')
  model.var('Z', desc='Autonomous expenditures')
  
  model.param('alpha_1', desc='Propensity to consume out of wages', default=0.8) # 0.8
  model.param('alpha_2', desc='Propensity to consume out of wealth', default=0.03375) # 0.2
  model.param('beta', desc='Expectation adjustment parameter', default=0.2) # 0.2
  model.param('beta_H', desc='Expectation adjustment parameter (Houses)', default=0.2) # 0.2
  model.param('beta_IN', desc='Expectation adjustment parameter (Inventories)', default=0.2) # 0.2
  model.param('gamma_0', desc='Autonomous component of residential investment', default=10) # 5
  model.param('gamma_1', desc='Sensitivity of residential investment of the houses own interest rate', default=0.01) # 15
  model.param('gamma_ph', desc='House pricing adjusment parameter', default=0.005) # 1
  model.param('gamma_Hp', desc='Sensitivity of houses supply for the lagged prices', default=0.5) # 1
  model.param('gamma_F', desc='% of distributed profits', default=0.4) # 0.3
  model.param('gamma_u', desc='Adjustment parameter for the marginal propensity to invest', default=0.014)
  model.param('gz', desc='Autonomous grouth rate', default=0.02)
  model.param('omegapar', desc='Wage-share', default=0.5)
  model.param('rm', desc='Interest rates on money deposits', default=0.02)
  model.param('spread_l', desc='Spread for loans', default=0.01)
  model.param('spread_mo', desc='Spread for mortgages', default=0.01)
  model.param('un', desc='Normal capacity utilization ratio', default=0.8)
  model.param('sigma_T', desc='Target inventories parameter', default = 0.1)
  model.param('v', desc='Capitl-Output ratio', default=2.5)
  model.param('x', desc='Stability corridor', default = 0.1)
  
  
  # General equations
  model.add('Y = C + I_t + d(IN)') # Eq1
  model.add('I_t = I_f + ph*I_h') # Eq2
  model.add('omega = omegapar') # Eq3
  model.add('W = omega*Y') # Eq4
  #model.add('Yk = K(-1)/v') # Eq5
  model.add('Yk = K_f/v') # Eq5
  model.add('u = Y/Yk') # Eq6
  model.add('gk = h*u/v') # Eq7
  model.add('Z = ph*I_h') # Eq8
  model.add('K = ph*H_S + K_f') # Eq9
  
  # Household equations
  model.add('YD = W + FD + rm*M(-1) - rmo*MO(-1)') # Eq10
  model.add('S_h = YD - C') # Eq11
  model.add('MO = ph*I_h + MO(-1)') # Eq12
  model.add('C = alpha_1*W + alpha_2*V_h(-1)') # Eq13
  model.add('V_h = M + ph*H_D - MO') # Eq14
  model.add('CG = d(ph)*H_D(-1) + ph(-1)*d(H_D)') # Eq15
  
  # Firms
  model.add('L = I_f - FU + d(IN) + L(-1)') # Eq16
  model.add('FT = Y - W') # Eq17
  model.add('FU = gamma_F*(FT - rl*L(-1))') # Eq18
  model.add('FD = (1 - gamma_F)*(FT - rl*L(-1))') # Eq19
  model.add('I_f = h*(Y)') # Eq20
  model.add('h = h(-1)*gamma_u*(u-un) + h(-1)') # Eq21 # Version without corridor
  model.add('d(K_f) = I_f')
  model.add('V_f = K_f + ph*IN*if_true(IN > 0) - L') # Eq22
  model.add('S_f = FU - I_f - d(IN)*ph') # Eq23
  
  # Banks
  model.add('M = (L - L(-1)) + (MO - MO(-1)) + M(-1)') # Eq24
  model.add('S_b = rl*L(-1) + rmo*MO(-1) - rm*M(-1)') # Eq25
  model.add('rl = rm + spread_l') # Eq26
  model.add('rmo = rm + spread_mo') # Eq27
  model.add('V_b = L + MO - M') # Eq28
  
  # Residential investment
  model.add('d(IN) = H_S - H_D') # Eq29
  #model.add('IN = (H_De - H_D + INe)*if_true(H_De - H_D + INe > 0)') # Eq29A
  #model.add('I_h = gamma_0 - gamma_1*rH') # Eq30
  model.add('I_h = (1 - rH)*MO(-1)*gamma_1') # Eq30
  model.add('d(H_D) = I_h') # Eq31
  model.add('rH = ((1 + rmo)/(1 + inf_ph)) - 1') # Eq32
  model.add('d(ph) = -gamma_ph*d(INe)*if_true(ph > 0)') # Eq33
  model.add('inf_ph = (d(ph)/ph(-1))*if_true(ph > 0)') # Eq34
  model.add('H_S = (1+g_Ih)*H_D(-1) + INe - IN(-1)') # Eq35
  model.add('H_De = H_De(-1) + beta_H*(H_D(-1) - H_De(-1))') # Eq36
  model.add('INe = IN(-1) + beta_IN*(INT - IN(-1))') # Eq37
  model.add('INT = sigma_T*H_De') # Eq38
  model.add('g_Ih = gz') # Eq39
  model.add('g_Ihe = g_Ihe(-1) + beta*(g_Ih(-1) - g_Ihe(-1))') # Eq40
  
  return model
```

# Solving

```
base = model()
df = SolveSFC(base, time=12)
df.transpose().round(2)
```

# Evaluating consistenty

```
evaldf = pd.DataFrame({
    'Households' : base.evaluate('M - MO + ph*H_D - V_h'),
    'Firms' : base.evaluate('K_f + ph*IN - L - V_f'),
    'Banks' : base.evaluate('L + MO - M - V_b'),
    '[Total Financial Wealth - K]' : base.evaluate('V_f + V_h + V_b - K'),
    #"Firm's Funds" : base.evaluate('I_f + d(IN) - FU - d(L)'),
    #"Housing" : base.evaluate('H_D - IN - H_S'),
    "[Saving - Investment]" : base.evaluate('S_f + S_b + S_h - ph*I_h - I_f'),
}, index = ['Sum'])
evaldf
```
