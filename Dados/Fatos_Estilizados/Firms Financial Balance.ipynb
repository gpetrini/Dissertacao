# Loading libraires

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas_datareader.data as web
import datetime

plt.style.use("seaborn-white")
%config InlineBackend.figure_format = 'retina'
```

```
start = datetime.datetime(1979, 1, 1)
end = datetime.datetime(2018, 1, 1)
```

Getting data

```
df = web.DataReader( # Nonfinancial Corporate Business
    ['TNWMVBSNNCB', # Net Worth, Level
    "TABSNNCB", # Total Assets, Level
    "TLBSNNCB", # Total Liabilities, Level
    ], 
    'fred', 
    start, 
    end)

df.columns = [
    "Net Worth",
    "Assets",
    "Liabilities"
]

df.head()
```

```
df["Net Worth/Liabilities"] = df["Net Worth"]/df["Liabilities"] - 1 
df["Assets/Liabilities"] = df["Assets"]/df["Liabilities"] - 1
df.head()
```

```
ax = df["Assets/Liabilities"].plot(color = "black")
ax.set_xlabel("")
ax.axhline(y = 0, color = "gray", ls = "--")
sns.despine()
plt.show()
```
