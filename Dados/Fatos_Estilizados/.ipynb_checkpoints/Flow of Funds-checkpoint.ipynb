# Importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dictionaty = pd.read_csv('z1-visualization/z1-visualization-dictionary.txt', sep="\t")

temp = []
level = []
series = []

for i in dictionaty['description']:
    level.append(i.split(';')[0])
    temp.append(i.split(';'))

for g in range(len(temp)):
    j = [len(temp) for temp in temp]
    if j[g] ==1:
        series.append(temp[g][0])
    else:
        series.append(temp[g][1])
for i in temp:
    if "Nonprofit" in i[0]:
        series[temp.index(i)] += '(Nonprofits)'

df = pd.read_csv('z1-visualization/z1-visualization-data.csv', index_col=0, parse_dates=True, dtype="float", names=series )
```
